## importing libraries and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales=pd.read_csv("D:\\machine learning\\course\\clustering\\K-means cluster\\Mall_Customers.csv")
x=sales.iloc[:,[3,4]].values
```

#### here we included only 2 column to train our model for better visualization of cluster we can include all of them too but this project is just for the implmentation and visualization of k-mean cluster

## Using elbow method to find optimal number of cluster we can make

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('elbow method class')
plt.xlabel("No of.cluster")
plt.ylabel('wcss')
plt.show()
```

#### no of optimal cluster we going to take here is 5 but we can take any depends on choice

## Training K-means clustering model on the dataset

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(x)
print(y_kmeans)
```

## visualzing the clusters

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
